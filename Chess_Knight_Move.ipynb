```python
start_point = "a1"
end_point   = "h8"
```

```python
def convert_square_to_number(square):
    s0 = ord(square[0]) - 96
    s1 = int(square[1])
    return s0*10 + s1

def convert_number_to_square(number):
    number = str(number)
    s0 = chr(int(number[0]) + 96)
    s1 = number[1]    
    return s0 + s1
```

```python
def all_moves(square):
    number = str(convert_square_to_number(square))
    moves = [-2, -1, 1, 2]
    new_moves = []
    for x in moves:
        for y in moves:
            if abs(x) + abs(y) == 3:
                new_x = int(number[0]) + x
                new_y = int(number[1]) + y
                if new_x > 0 and new_x < 9 and new_y > 0 and new_y < 9:
                    new_number = new_x* 10 + new_y
                    #print(convert_number_to_square(new_number))
                    new_moves.append(convert_number_to_square(new_number))
    return new_moves
```

```python
def show_chain(chain):
    return[convert_number_to_square(move) for move in chain]
#print(*show_chain(matrix[0]), sep = "-")
```

```python
%%time
print("Start point:", start_point, "End point:", end_point)
matrix = {0:[convert_square_to_number(start_point)]}
flag = False
all_variants = 0

for step in range(25):
    new_matrix = {}
    count = 0
    for i in matrix:
        chain = list(matrix[i])
        #print("chain:",chain, i, "len",len(chain))
        last_move = chain[len(chain) - 1]
        #print("last_move:", last_move, convert_number_to_square(last_move))
        
        new_moves = all_moves(convert_number_to_square(last_move))
        for move in new_moves:
            new_chain_dig = list(chain)
            new_chain_dig.append(convert_square_to_number(move))
            #print("chain", chain, " move", move, convert_square_to_number(move), "count", count,"new_chain_dig", new_chain_dig, show_chain(new_chain_dig))
            #print(*show_chain(new_chain_dig), sep = "-")
            
            new_matrix[count] = new_chain_dig
            count += 1
            
            if move == end_point:
                all_variants += 1
                print("#"* 20, "Chain len:", len(new_chain_dig) - 1)
                print(*show_chain(new_chain_dig), sep = "-")
                flag = True
            
    matrix = dict(new_matrix)
    if flag: print("ALL Variants = ", all_variants);break
```
